```python

import matplotlib.pyplot as plt
import numpy as np

trajs = [traj1, traj2, traj3, traj4,traj5,traj6,traj7,traj8,traj9,traj10,traj11,traj12,traj13,traj14,traj15,traj16]

spheres = []
Prin_x = []
Prin_y = []
Prin_z = []
ACs = []

for i, traj in enumerate(trajs):
    sphere = md.asphericity(traj)
    Prin = np.squeeze(md.principal_moments(traj))
    AC = md.acylindricity(traj)
    spheres.append(sphere)
    Prin_x.append(Prin[0])
    Prin_y.append(Prin[1])
    Prin_z.append(Prin[2])
    ACs.append(AC)
    print(f"Sphere {i} is: {sphere} and PM {i} is: {Prin}")
    print(f"Cylindricity of {i} is: {AC}")

# Plotting
plt.figure(figsize=(10, 6))

# Define custom x-axis values and labels
x_values = [0, 1, 2, 3]
x_labels = ['.5', '1', '2', '5']

plt.subplot(2, 2, 1)
plt.plot(range(len(spheres)), spheres, marker='o', linestyle='')
plt.xlabel('Epsilon')
plt.ylabel('Sphericity')
plt.title('Sphericity of Structures')
plt.xticks(x_values, x_labels) 

plt.subplot(2, 2, 2)
plt.plot(range(len(Prin_x)), Prin_x, label='X', marker='o', linestyle='')
plt.plot(range(len(Prin_y)), Prin_y, label='Y', marker='o', linestyle='')
plt.plot(range(len(Prin_z)), Prin_z, label='Z', marker='o', linestyle='')
plt.xlabel('Epsilon')
plt.ylabel('Principal Moments')
plt.title('Principal Moments of Structures')
plt.legend()
plt.xticks(range(len(Prin))) 
plt.xticks(x_values, x_labels) 

plt.subplot(2, 2, 3)
plt.plot(range(len(ACs)), ACs, marker='o', linestyle='')
plt.xlabel('Epsilon')
plt.ylabel('Acylindricity')
plt.title('Acylindricity of Structures')
plt.xticks(x_values, x_labels) 
#plt.xticks(range(len(spheres))) 

plt.tight_layout()
plt.show()
```
